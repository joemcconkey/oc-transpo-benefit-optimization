Linearized formulation (reference, r02)

Goal: replace the nonlinear demand and benefit expressions with their first-order Taylor expansions about the baseline service level.

Decision is number of buses per route per time block

Sets / indices

$$
r \in R \quad \text{(routes)}, \qquad t \in T \quad \text{(time blocks)}
$$

Decision variable

$$
n_{r,t}^{\text{new}} \in \mathbb{Z}
$$

number of buses assigned to route \(r\) at time \(t\)

Given parameters

$$
n_{r,t}^{\text{old}}, \quad x_{r,t}^{\text{old}}, \quad \varepsilon
$$

$$
\alpha_{r,t}, \quad \beta_{r,t}^{\text{time}}, \quad \beta_{r,t}^{\text{em}}
$$

$$
c_{r,t}^{\text{lab}}, \quad c_{r,t}^{\text{fuel}}, \quad c_{r,t}^{\text{maint}}
$$

Change in buses

$$
\Delta n_{r,t} = n_{r,t}^{\text{new}} - n_{r,t}^{\text{old}}
$$

Costs (linear)

$$
\Delta C_{r,t} =
c_{r,t}^{\text{lab}} \, \Delta n_{r,t}
+ c_{r,t}^{\text{fuel}} \, \Delta n_{r,t}
+ c_{r,t}^{\text{maint}} \, \Delta n_{r,t}
$$

Demand (first-order Taylor linearization)

Nonlinear reference model:

$$
x_{r,t}^{\text{new}}(n)=x_{r,t}^{\text{old}}\left(\frac{n}{n_{r,t}^{\text{old}}}\right)^{\varepsilon}
$$

Taylor expansion about \(n=n_{r,t}^{\text{old}}\):

$$
x_{r,t}^{\text{new}}(n)\approx x_{r,t}^{\text{old}} + \alpha_{r,t}\,\Delta n_{r,t}
$$

with

$$
\alpha_{r,t}
=
\left.\frac{d x_{r,t}^{\text{new}}}{dn}\right|_{n=n_{r,t}^{\text{old}}}
=
\varepsilon \frac{x_{r,t}^{\text{old}}}{n_{r,t}^{\text{old}}}
$$

Benefits (first-order Taylor linearization)

Time-benefit term:

$$
\Delta B_{r,t}^{\text{time}} \approx \beta_{r,t}^{\text{time}} \, \Delta n_{r,t}
$$

with

$$
\beta_{r,t}^{\text{time}}
=
\left.\frac{d\,\Delta B_{r,t}^{\text{time}}}{dn}\right|_{n=n_{r,t}^{\text{old}}}
=
\frac{x_{r,t}^{\text{old}}\,W_{\text{passenger}}\,F_{\text{wage},t}\,T_{r,t}}{2\,(n_{r,t}^{\text{old}})^2}
$$

Emissions-benefit term:

$$
\Delta B_{r,t}^{\text{em}} \approx \beta_{r,t}^{\text{em}} \, \Delta n_{r,t}
$$

with

$$
\beta_{r,t}^{\text{em}}
=
\left.\frac{d\,\Delta B_{r,t}^{\text{em}}}{dn}\right|_{n=n_{r,t}^{\text{old}}}
=
\alpha_{r,t}\,\frac{L_{r}^{\text{avg trip}}\,I_{\text{GHG,car}}\,SCC}{1000}
$$

Total linearized benefit:

$$
\Delta B_{r,t} =
\left(\beta_{r,t}^{\text{time}} + \beta_{r,t}^{\text{em}}\right)\Delta n_{r,t}
$$

Objective

$$
\min \sum_{r,t} \left( \Delta C_{r,t} - \Delta B_{r,t} \right)
$$

linear in \(n_{r,t}^{\text{new}}\)

Constraints

Fleet:

$$
\sum_{r} n_{r,t}^{\text{new}} \leq \text{fleet}_t
$$

Budget:

$$
\sum_{r,t} \Delta C_{r,t} \leq \text{budget}
$$

Bounds:

$$
n_{r,t}^{\min} \leq n_{r,t}^{\text{new}} \leq n_{r,t}^{\max}
$$

Integrality:

$$
n_{r,t}^{\text{new}} \in \mathbb{Z}
$$

What was simplified

Original nonlinear model had curvature such as

$$
\text{wait time} \sim \frac{1}{n_{r,t}^{\text{new}}}, \qquad
x_{r,t}^{\text{new}} \sim \left(\frac{n_{r,t}^{\text{new}}}{n_{r,t}^{\text{old}}}\right)^{\varepsilon}
$$

These are replaced by first-order Taylor expansions in \(\Delta n_{r,t}\) around the baseline service level \(n_{r,t}^{\text{old}}\).

Interpretation:

constant local marginal benefit per added bus, evaluated at the current service level

Tradeoff:

simpler MILP, but diminishing returns are only captured locally near the baseline
