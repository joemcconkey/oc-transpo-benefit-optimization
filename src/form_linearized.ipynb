Linearized formulation (reference)

Goal: simplify nonlinear demand + benefit terms so model is easy to solve with MILP (Gurobi)

Decision is number of buses per route per time block

Sets / indices

$$
r \in R \quad \text{(routes)}, \qquad t \in T \quad \text{(time blocks)}
$$

Decision variable

$$
n_{r,t}^{\text{new}} \in \mathbb{Z}
$$

number of buses assigned to route \(r\) at time \(t\)

Given parameters

$$
n_{r,t}^{\text{old}}, \quad x_{r,t}^{\text{old}}
$$

$$
\beta_{r,t}^{\text{time}}, \quad \beta_{r,t}^{\text{em}}
$$

$$
c_{r,t}^{\text{lab}}, \quad c_{r,t}^{\text{fuel}}, \quad c_{r,t}^{\text{maint}}
$$

Change in buses

$$
\Delta n_{r,t} = n_{r,t}^{\text{new}} - n_{r,t}^{\text{old}}
$$

Costs (linear)

$$
\Delta C_{r,t} =
c_{r,t}^{\text{lab}} \, \Delta n_{r,t}
+ c_{r,t}^{\text{fuel}} \, \Delta n_{r,t}
+ c_{r,t}^{\text{maint}} \, \Delta n_{r,t}
$$

Demand (linearized)

$$
x_{r,t}^{\text{new}} =
x_{r,t}^{\text{old}} + \alpha_{r,t} \, \Delta n_{r,t}
$$

no elasticity / no square-root dependence

Benefits (linearized)

$$
\Delta B_{r,t}^{\text{time}} = \beta_{r,t}^{\text{time}} \, \Delta n_{r,t}
$$

$$
\Delta B_{r,t}^{\text{em}} = \beta_{r,t}^{\text{em}} \, \Delta n_{r,t}
$$

$$
\Delta B_{r,t} =
\left(\beta_{r,t}^{\text{time}} + \beta_{r,t}^{\text{em}}\right) \Delta n_{r,t}
$$

Objective

$$
\min \sum_{r,t} \left( \Delta C_{r,t} - \Delta B_{r,t} \right)
$$

linear in \(n_{r,t}^{\text{new}}\)

Constraints

Fleet:

$$
\sum_{r} n_{r,t}^{\text{new}} \leq \text{fleet}_t
$$

Budget:

$$
\sum_{r,t} \Delta C_{r,t} \leq \text{budget}
$$

Bounds:

$$
n_{r,t}^{\min} \leq n_{r,t}^{\text{new}} \leq n_{r,t}^{\max}
$$

Integrality:

$$
n_{r,t}^{\text{new}} \in \mathbb{Z}
$$

What was simplified

Original model had:

$$
\text{wait time} \sim \frac{1}{n_{r,t}^{\text{new}}}, \quad
x_{r,t}^{\text{new}} \sim \left(\frac{n_{r,t}^{\text{old}}}{n_{r,t}^{\text{new}}}\right)^{0.5}
$$

Both replaced with linear relationships in 

$$
\Delta n_{r,t}
$$

Interpretation:

constant marginal benefit per added bus

Tradeoff:

simpler MILP, no diminishing returns
